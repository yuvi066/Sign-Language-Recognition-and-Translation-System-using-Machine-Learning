# Feature Extraction

```python
def extraction(results):
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    poses = np.array([[res.x,res.y,res.z,res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(132)
    lh = np.array([[res.x,res.y,res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(63)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(63)
    # faces = np.array([[res.x,res.y,res.z] for res in results.
    return np.concatenate([poses, lh , rh])
    
```

```python
def draw(image , results):
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION)
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks,mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks,mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image, results.pose_landmarks,mp_holistic.POSE_CONNECTIONS)
```

```python
def detectVideo(image, model):
    results = model.process(image)
    return image, results
```

# calculating frames to drop

```python
def frame_to_skip(size, threshold):
    for i in range(1 , size):
        to_skip = size // i 
        if size - to_skip <= threshold:
            continue
        else:
            return (i - 1)
    return 0
```

```python
def dry_run(frame_size):
    count = 0
    filtered = {}
    for val in frame_size: 
        to_skip = frame_to_skip(frame_size[val] , 30)
        result = frame_size[val]
        if to_skip > 0:
            result = frame_sizes[val] - (frame_size[val] // to_skip)
        if result == 0:
            count+=1
        print(f'initial = {frame_size[val]} ----> {result} ({to_skip}) ({val})')
        if(result != 0):
            filtered[val] = [to_skip, result]
    print(count)
    return filtered
    
```

```python
def detect_camera(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # COLOR CONVERSION BGR 2 RGB
    image.flags.writeable = False                  # Image is no longer writeable
    results = model.process(image)                 # Make prediction
    image.flags.writeable = True                   # Image is now writeable 
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # COLOR COVERSION RGB 2 BGR
    return image, results
```
